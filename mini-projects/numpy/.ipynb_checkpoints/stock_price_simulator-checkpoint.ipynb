```python
import numpy as np
import matplotlib.pyplot as plot

starting_price = 100
num_steps = 100
price_changes = np.random.rand(num_steps) * 2 - 1
stock_prices = np.cumsum(price_changes) * starting_price + starting_price

plot.plot(stock_prices)
plot.xlabel("Time Step")
plot.ylabel("Stock Price")
plot.title("Simulated Stock Price Movements")
plot.show()
```
