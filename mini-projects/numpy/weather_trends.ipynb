```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
temparature = np.random.rand(100) * 30
humidity = np.random.rand(100) * 100
```

```python
plot.scatter(temparature, humidity, color="green", marker="s")
plot.grid(True)
plot.xlabel("Temparature (C)")
plot.ylabel("Humidity (%)")
plot.title("Weather Trends")
plot.show()
```
